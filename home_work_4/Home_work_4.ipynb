Задание № 2

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, color='green')

plt.title('График f(t)')

plt.xlabel('Значения t')

plt.ylabel('Значения f')

plt.axis([0.5, 9.5, -2.5, 2.5])

plt.show()
```

```python
x = np.linspace(-3, 3, 51)
x

y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x -1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 =ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2')

ax3.plot(x, y3)
ax3.set_title('График y3')

ax4.plot(x, y4)
ax4.set_title('Графиу y4')

plt.show()
```

Задание № 4

```python

bmi_df = pd.read_csv('creditcard.csv')

plt.style.use('fivethirtyeight')


bmi_df_1 = bmi_df['Class'].value_counts()
bmi_df_1.plot(kind='barh')

plt.show()
```

```python
bmi_df_1.plot(kind='bar', logy=True)

plt.show()
```

```python
from pylab import rcParams
bmi_1 = bmi_df.loc[bmi_df['Class'] == 0, ['V1']]
bmi_2 = bmi_df.loc[bmi_df['Class'] == 1, ['V1']]
```

```python

plt.hist(bmi_1['V1'], bins=20, alpha=0.5, density=True, label='Class0', color='darkgrey', ec='black')
plt.hist(bmi_2['V1'], bins=20, alpha=0.5, density=True, label='Class1', color='red', ec='white')

rcParams["figure.figsize"] = 10, 12

plt.legend()
plt.show()
```
